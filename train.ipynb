# Train

```python
import numpy as np
import os
import pandas as pd
import gradient_descent

mnist_train = pd.read_csv(os.path.join('data', 'mnist_train.csv'))
mnist_train = np.array(mnist_train)

data_train = mnist_train.T
Y_train = data_train[0]
X_train = data_train[1: mnist_train.shape[1]] / 255

weights_and_biases = gradient_descent.gradient_descent(X_train, Y_train, 10000, 0.01, layers=2)

parameters = {}
for index in range(len(weights_and_biases)):
    W, b = weights_and_biases[index]
    parameters[f'W{index}'] = W
    parameters[f'b{index}'] = b
np.savez(os.path.join('data', 'trained_params'), **parameters)
```
